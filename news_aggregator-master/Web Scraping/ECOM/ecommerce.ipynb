```python
from bs4 import BeautifulSoup
from joblib import Parallel, delayed
import requests
import pandas as pd 
import time
```

```python
base_url = 'https://books.toscrape.com/catalogue/'

product_links = set()

for x in range(1,51):
    r = requests.get(f'https://books.toscrape.com/catalogue/page-{x}.html')
    soup = BeautifulSoup(r.text, 'html.parser')
    
    product_list=soup.find_all('article',class_='product_pod')
    
    for item in product_list:
        for link in item.find_all('a', href=True):
            product_links.add(base_url + link['href'])

product_links = list(product_links)
```

```python
product_data=[]
for link in product_links:
    r=requests.get(link)
    soup=BeautifulSoup(r.text,'lxml')
    name=soup.find('h1').text.strip()
    availability=soup.find('p',class_='instock availability').text.strip()
    price=soup.find('p').text.strip('')
    rating = soup.find('p', class_='star-rating')['class'][1] 

    Book={'name':name,'availablity':availability,'pirce':price,'rating':rating}
    print(Book)
    product_data.append(Book)
```

```python
data=pd.DataFrame(product_data)
```

```python
data.to_csv('E-commerce_scraped_data.csv',index=False)
```

```python
def search_product_df(user_query, data):
    # Case-insensitive search for 'name' column with partial matches
    results = data[data['name'].str.contains(user_query, case=False)]
    return results
while True:
    search_input = input("Enter a search query (or 'exit' to quit): ")
    if search_input.lower() == 'exit':
        break
    search_results_df = search_product_df(search_input, data)
    if search_results_df.empty:
        print("No results found.")
    else:
        print(search_results_df)
```
