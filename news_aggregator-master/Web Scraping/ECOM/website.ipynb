```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 
```

```python
source=requests.get('https://www.scrapethissite.com/pages/forms/').text
soup=BeautifulSoup(source,'lxml')
```

```python
teams = soup.find_all('tr', class_='team')
```

```python
teams
```

```python

scraped_data = []
for team in teams:
    team_data = {}
    team_data['name'] = team.find('td', class_='name').text.strip()
    team_data['year'] = team.find('td', class_='year').text.strip()
    team_data['wins'] = team.find('td', class_='wins').text.strip()
    team_data['losses'] = team.find('td', class_='losses').text.strip()
    team_data['ot_losses'] = team.find('td', class_='ot-losses').text.strip()
    team_data['win_percentage'] = team.find('td', class_='pct').text.strip()
    team_data['goals_for'] = team.find('td', class_='gf').text.strip()
    team_data['goals_against'] = team.find('td', class_='ga').text.strip()
    team_data['plus_minus'] = team.find('td', class_='diff').text.strip()
    scraped_data.append(team_data)
for data in scraped_data:
    print(data)
```

```python
df=pd.DataFrame(scraped_data)
```

```python
df.to_csv('scraped_data_2.csv', index=False)
```
