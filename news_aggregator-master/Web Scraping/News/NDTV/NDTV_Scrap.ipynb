```python
from bs4 import BeautifulSoup
import requests
```

```python
#Test 
url='https://www.ndtv.com/india/page-1'
page=requests.get(url)
soup=BeautifulSoup(page.text,'lxml')
```

```python
soup.find_all('h2')
```

```python
#test
for h2_tag in soup.find_all('h2'):
    headline = h2_tag.text.strip()
    link = h2_tag.a['href']
    news_paragraphs = soup.find_all('p')
    news_text = '\n'.join([p.text.strip() for p in reversed(news_paragraphs)])
    print(headline)
    print(news_text)
    print(link)
    print()
```

```python
India_News_Data = set()

for x in range(1, 15):
    r = requests.get(f'https://www.ndtv.com/india/page-{x}')
    soup = BeautifulSoup(r.text, 'lxml')
    
    News_Data = soup.find_all('h2')
    
    for h2_tag in News_Data:
        headline = h2_tag.text.strip()
        link = h2_tag.a['href']
        news_paragraphs = soup.find_all('p')[:2]
        news_text = '\n'.join([p.text.strip() for p in reversed(news_paragraphs)])
        print(headline)
        print(news_text)
        print(link)
        print()

     
```
