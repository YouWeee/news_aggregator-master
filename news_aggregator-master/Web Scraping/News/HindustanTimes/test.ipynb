```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("--headless")  

driver = webdriver.Chrome(options=chrome_options)

url_template = 'https://www.hindustantimes.com/india-news/page-{page_number}'

all_headlines = []
all_links = []

for page_number in range(1, 3):  
    url = url_template.format(page_number=page_number)

    driver.get(url)
    driver.implicitly_wait(10)  
    page_source = driver.page_source

    soup = BeautifulSoup(page_source, 'html.parser')
    

    h3_tags = soup.find_all('h3')
    
    

    for h3 in h3_tags:
        a_tag = h3.a
        if a_tag:
            # Extract href and headline if the <a> tag exists
            base_url='https://www.hindustantimes.com'

            link = base_url+a_tag.get('href')
            headline = a_tag.text
            all_headlines.append(headline)
            all_links.append(link)

# Close the Chrome driver
driver.quit()

base_url='https://www.hindustantimes.com'
# Print the extracted headlines and links
for headline, link in zip(all_headlines, all_links):
    print("Headline:", headline)
    print("Link:",link)
    print("---")
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("--headless")  

driver = webdriver.Chrome(options=chrome_options)

url_template = 'https://www.hindustantimes.com/india-news/page-{page_number}'

for page_number in range(1, 2):  
    url = url_template.format(page_number=page_number)

    driver.get(url)

    driver.implicitly_wait(10)  

    page_source = driver.page_source

    
    soup = BeautifulSoup(page_source, 'html.parser')

    
    h3_tags = soup.find_all('h3')

   
    for h3 in h3_tags:
        
        a_tag = h3.a
        if a_tag:
           
            base_url = 'https://www.hindustantimes.com'
            link = base_url + a_tag.get('href')
            headline = a_tag.text

            driver.get(link)
            driver.implicitly_wait(10)  # Wait for 10 seconds

            article_page_source = driver.page_source

            article_soup = BeautifulSoup(article_page_source, 'html.parser')

            sortdec_tag = article_soup.find('h2', class_='sortDec') 
            sortdec_text = sortdec_tag.text if sortdec_tag else "Sortdec not found"

            print("Headline:", headline)
            print("Sortdec:", sortdec_text)
            print("Link:", link)
            print("---")

driver.quit()
```
