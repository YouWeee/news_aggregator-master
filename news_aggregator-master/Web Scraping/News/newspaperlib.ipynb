```python
import newspaper

bbc = newspaper.build('https://www.aljazeera.com/news/')

# Get titles and summaries of all articles
for article in bbc.articles:
    article.download()
    article.parse()
    article.nlp()
    
    # Check if the article is in English
    if article.meta_lang == 'en':
        print("Title:", article.title)
        print("Summary:", article.summary)
        print()
```
